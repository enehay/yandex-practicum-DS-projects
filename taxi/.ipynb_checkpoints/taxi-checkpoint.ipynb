#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Мне нужно построить модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Для этого я:

1. Загружу данные и выполню их ресемплирование по одному часу.
2. Проанализирую данные.
3. Обучу разные модели с различными гиперпараметрами. Сделаю тестовую выборку размером 10% от исходных данных.
4. Проверю данные на тестовой выборке и сделаю выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Подготовка

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
pd.set_option('display.max_rows', None) #снимает ограничение на отображение количества строк
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.metrics import make_scorer
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import TimeSeriesSplit
```

```python
data = pd.read_csv('/datasets/taxi.csv', index_col=[0], parse_dates=[0])
data.sort_index(inplace=True) #cjhnbhjdrf
print(f"\n Информация об обучающей выборке \n")
display(data.info())
display(data.describe())
print(data.shape)
print(f"\n Явных дубликатов", data.duplicated().sum())
display(data.head(7))
```

```python
data = data.resample('1H').sum() # ресемплирование по часу
data.plot(figsize = (15,10))
```

## Анализ

Чтобы лучше понимать временной ряд, разберём тренды и сезонность.

```python
#произведем декомпозицию временного ряда 

decomposed = seasonal_decompose(data)
plt.figure(figsize=(15,10))
plt.subplot(311)
# Чтобы график корректно отобразился, указываем его
# оси ax, равными plt.gca() (англ. get current axis,
# получить текущие оси)
decomposed.trend.plot(ax=plt.gca())
plt.title('Trend')
plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca()) 
plt.title('Seasonality')
plt.subplot(313)
decomposed.resid.plot(ax=plt.gca())
plt.title('Residuals')
plt.tight_layout()
```

Из графика тренда видно, что ряд нестационарный, потому что у него увеличивается среднее значение суммы заказов, особенно это видно в конце августа.
Также видно, что разброс значений ряда существенно варьируется в зависимости от периода - дисперчия увеличивается.

Сделаем срез за август, чтобы увидеть тренд, сезонность, шум более наглядно.

```python
data_2 = data['2018-08-22':'2018-08-31']
decomposed_2 = seasonal_decompose(data_2)
plt.figure(figsize=(15,10))
plt.subplot(311)
# Чтобы график корректно отобразился, указываем его
# оси ax, равными plt.gca() (англ. get current axis,
# получить текущие оси)
decomposed_2.trend.plot(ax=plt.gca())
plt.title('Trend')
plt.subplot(312)
decomposed_2.seasonal.plot(ax=plt.gca()) 
plt.title('Seasonality')
plt.subplot(313)
decomposed_2.resid.plot(ax=plt.gca())
plt.title('Residuals')
plt.tight_layout()
```

В течение дня заметна сезонность

Чтобы во временном ряду снизить колебания, применим скользящее среднее.

Нам нужно будет спрогнозировать данные на следующий час, поэтому возьмем скользящее среднее = 24 за сутки

```python
data['rolling_mean'] = data['num_orders'].shift().rolling(24).mean()
data.plot(figsize = (15,10))
```

```python
data['num_orders'].shift().rolling(24).mean()
```

## Обучение

Создадим признаки для горизонта прогнозирования

```python
def make_features(data, max_lag, rolling_mean_size):
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour 
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()

make_features(data, 100, 100)
data.info()
print(data.head())
```

```python
train, test = train_test_split(data, shuffle=False, test_size=0.1)
train = train.dropna()
print(f"Размер обучающей выборки: {train.shape}, размер тестовой выборки: {test.shape}")
```

```python
features_train  = train.drop('num_orders', axis=1)
target_train = train['num_orders']
features_test = test.drop('num_orders', axis=1)
target_test = test['num_orders']
```

```python
print(features_train.shape, target_train.shape, features_test.shape, target_test.shape)
```

```python
def rmse(target,predict):
     return ((mean_squared_error(target,predict))**0.5)
# neg_mean_squared_error(target,predict).mean() * -1) ** 0.5
# {mean_squared_error(target_valid_0, predictions_valid_0)**0.5}
```

### Линейная регрессия

```python
from timeit import default_timer as timer
```

```python
%%time
start = timer()
model_lr = LinearRegression()
model_lr.fit(features_train, target_train) #обучение
rmse_lr = cross_val_score(model_lr,features_train, target_train, 
                         cv = TimeSeriesSplit(n_splits=24), 
                         n_jobs = -1, 
                         scoring = 'neg_root_mean_squared_error').mean().round(2)*(-1)
print(f'''Значение метрики RMSE на обучающей выборке для LinearRegression''', rmse_lr)
#повторение процесса разбиения временного ряда на 24 раза
end = timer()
tl_2 = end - start
print('Затрачено времени на обучение',round(tl_2, 3))
```

### Дерево решений

```python
%%time
model_3 = DecisionTreeRegressor(random_state = 12345)
pipelin = Pipeline(steps=[("model", model_3)])
param_grid = {"model__max_depth":range(1,101,10)} 
search_3 = GridSearchCV(pipelin, param_grid, scoring = 'neg_root_mean_squared_error',  n_jobs=-1, cv=TimeSeriesSplit(n_splits=24), verbose=1)
search_3.fit(features_train, target_train)
z = search_3.best_score_
print(search_3.best_params_, -1*z)
```

```python
%%time
start = timer()
model_3 = DecisionTreeRegressor(max_depth = 11, random_state = 12345)
model_3.fit(features_train, target_train)
end = timer()
tl_11 = end - start
print('Затрачено времени на обучение',round(tl_11, 3))
```

```python
%%time
start = timer()
target_predict = model_3.predict(features_train)
end = timer()
tl_12 = end - start
print('Затрачено времени на предсказание',round(tl_12, 3))
```

```python
z = rmse(target_train,target_predict).round(2)
print("\nrmse: ",z)
```

### CatBoost (от англ. categorical boosting; «бустинг для категориальных признаков»)

```python
%%time
start = timer()
model_1 = CatBoostRegressor()

params = [{'learning_rate': [.03, .1],
        'depth': [4, 6, 10],
        'iterations': [1, 15,5]}]

grid_search_result = model_1.grid_search(params,
                                       X=features_train,
                                       y=target_train,
                                       cv=TimeSeriesSplit(n_splits=24),
                                       plot=False)
end = timer()
tl_5 = end - start
print('Затрачено времени на исполнение',round(tl_5, 3))
print("\nBest Params : ", grid_search_result['params'])
```

```python
%%time
start = timer()
model_1 = CatBoostRegressor(loss_function='RMSE', iterations=15, depth = 4,learning_rate = .1, ) 
model_1.fit(features_train, target_train) #время обучения
end = timer()
tl_6 = end - start
print('Затрачено времени на обучения',round(tl_6, 3))
```

```python
%%time
start = timer()
predict_cb=model_1.predict(features_train)
end = timer()
tl_7 = end - start
print('Затрачено времени на предсказание',round(tl_7, 3))
```

```python
y = rmse(target_train, predict_cb)
print(y) 
```

### LightGBM

```python
%%time

model_2 = LGBMRegressor(random_state = 12345) 

params = [{'num_leaves':[4, 6, 10], 'learning_rate':[.1, .3, .5]}]

grid_search_result_2 = GridSearchCV(model_2, params, scoring='neg_root_mean_squared_error', cv = TimeSeriesSplit(n_splits=8))
grid_search_result_2.fit(features_train, target_train)
a = grid_search_result_2.best_score_

print(grid_search_result_2.best_params_, -1*a)
```

```python
%%time
start = timer()
model_2 = LGBMRegressor(num_leaves = 10, learning_rate = .1, random_state = 12345)
model_2.fit(features_train, target_train)
end = timer()
tl_9 = end - start
print('Затрачено времени на обучение',round(tl_9, 3))
```

```python
%%time
start = timer()
predict_lg=model_2.predict(features_train)
end = timer()
tl_10 = end - start
print('Затрачено времени на предсказание',round(tl_10, 3))
```

```python
g = rmse(target_train, predict_lg)
print(g) 
```

```python
result = pd.DataFrame(
    [[rmse_lr,z,y,g]], 
    columns=['LinearRegression', 'DecisionTreeRegressor', 'CatBoostRegressor', 'LightGBM'], 
    index=['RMSE']
)
result
```

## Тестирование

```python
model_2 = LGBMRegressor(num_leaves = 10, learning_rate = .1, random_state = 12345) 
model_2.fit(features_train, target_train)
target_predict = model_2.predict(features_test)
print("\nrmse: ",rmse(target_test,target_predict).round(2))
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(target_test.index, target_test, label='Заказы в тестовой выборке')
ax.plot(target_test.index, target_predict, label='Заказы предсказанной выборки')
plt.title('Сравнение реальных заказов и предсказанных значений заказов')
plt.legend()
plt.show()
```

Выбранная модель хорошо справляется с задачей предсказывать количество заказов с разбивкой по часам, что отображено на графике.
Лучшую rmse на трейне показала модель LGBMRegressor, на тесте rmse этой модели составило 38,7
